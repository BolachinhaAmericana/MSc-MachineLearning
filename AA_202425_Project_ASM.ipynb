### AdoptionSpeed

The value of `AdoptionSpeed` describes how quickly, if at all, a pet is adopted:
* 0 - Pet was adopted on the same day as it was listed. 
* 1 - Pet was adopted between 1 and 7 days (1st week) after being listed. 
* 2 - Pet was adopted between 8 and 30 days (1st month) after being listed. 
* 3 - Pet was adopted between 31 and 90 days (2nd & 3rd month) after being listed.
* 4 - No adoption after 100 days of being listed. (There are no pets in this dataset that waited between 90 and 100 days).

```python
def plot_bar_with_counts(series, title, labels=None):
    """
    Creates a bar plot with counts displayed on top of each bar.

    Parameters:
    - series: Pandas Series, the data to plot.
    - title: str, the title of the plot.
    - labels: dict, optional, mapping of series values to custom labels for the x-axis.

    Returns:
    - None: Displays the plot.
    """
    # Ensure the input is a pandas Series
    if not isinstance(series, pd.Series):
        raise TypeError("Input must be a pandas Series.")
    
    # Create a DataFrame for plotting
    data = pd.DataFrame({"Category": series})

    order = sorted(series.unique())
    
    # Map labels if provided
    if labels:
        data["Category"] = data["Category"].map(labels)
        order = [labels[val] for val in order]  # Apply label mapping to the order
    
    # Plot bar chart
    plt.figure(figsize=(8, 6))
    ax = sns.countplot(x="Category", data=data, palette="pastel", order=order)
    
    # Add count labels to bars
    ax.bar_label(ax.containers[0])
    
    # Set plot labels and title
    plt.ylabel("Count", fontsize=14)
    plt.title(title, fontsize=16)
    plt.xticks(fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
def plot_countplot(series1, series2, labels1=None, labels2=None, title=None):
    """
    Plots a countplot for two categorical variables, sorted by their values.

    Parameters:
    - series1: Pandas Series, first categorical variable (e.g., "Type").
    - series2: Pandas Series, second categorical variable (e.g., "IsAdopted").
    - labels1: dict, optional, mapping of values in series1 to labels (e.g., {1: "Dog", 2: "Cat"}).
    - labels2: dict, optional, mapping of values in series2 to labels (e.g., {0: "Not Adopted", 1: "Adopted"}).
    - title: str, optional, title for the plot.

    Returns:
    - None: Displays the plot.
    """
    # Ensure inputs are pandas Series
    if not isinstance(series1, pd.Series) or not isinstance(series2, pd.Series):
        raise TypeError("Both inputs must be pandas Series.")
    
    # Create a DataFrame for plotting
    data = pd.DataFrame({ "Series1": series1, "Series2": series2 })
    
    # Sort unique values for series1 and series2
    order1 = sorted(series1.unique())
    order2 = sorted(series2.unique())
    
    # Map labels if provided
    if labels1:
        data["Series1"] = data["Series1"].map(labels1)
        order1 = [labels1[val] for val in order1]  # Apply label mapping to the order
    if labels2:
        data["Series2"] = data["Series2"].map(labels2)
        order2 = [labels2[val] for val in order2]  # Apply label mapping to the order
    
    # Plot countplot
    plt.figure(figsize=(8, 6))
    ax = sns.countplot(
        x="Series1",
        hue="Series2",
        data=data,
        palette="pastel",
        order=order1,
        hue_order=order2
    )
    
    # Add count labels to bars
    for container in ax.containers:
        ax.bar_label(container)
    
    # Set plot labels and title
    plt.xlabel(series1.name, fontsize=14)
    plt.ylabel("Count", fontsize=14)
    if title:
        plt.title(title, fontsize=16)
    plt.legend(title=series2.name, fontsize=12, title_fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
def two_prop_test(x1, x2, n1, n2, alternative):

    count = np.array([x1, x2])
    nobs = np.array([n1, n2])
    print(count)
    print(nobs)
    print(count/nobs)

    z_stat, p_value = proportions_ztest(count, nobs, alternative=alternative)

    # Output results
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")    
```

Podemos testar as seguintes hipóteses H0: p1 = p2 versus H1: p1 ≠ p2
Ou seja, a hipótese nula de que a probabilidade de um perfil de cães resultar em adoção é igual à probabilidade de um perfil de gatos resultar em adoção contra a hipótese alternativa de que estas probabilidades são diferentes.

```python
two_prop_test(x1 = df[["Type", "IsAdopted"]].value_counts()[1,1],
              x2 = df[["Type", "IsAdopted"]].value_counts()[2,1],
              n1 = df["Type"].value_counts()[1],
              n2 = df["Type"].value_counts()[2],
              alternative= "two-sided")
```

Atendendo ao p-values, rejeita-se a hipótese nula para os níveis de significância habituais (0.1, 0.05, 0.01). Ou seja, rejeita-se a hipótese de que p1 = p2. Existe evidência estatística de que um perfil de gatos tem uma maior probabilidade em resultar em adoção.
